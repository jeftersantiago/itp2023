# Simulação: Problema de Monty Hall

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from random import seed
seed(100) 
```

```python
doors = [1, 2, 3]
N = 5000
wins = []
games = []

for i in np.arange(1, N, 2):
  win = 0
  lost = 0
  for j in range(0, i):
    C = random.choice(doors)
    X = random.choice(doors)
    if(C != X):
      win += 1
  wins.append(win/i)
  games.append(i)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(games, wins, '-', label = "Simulação: Muda a escolha de porta", color = "green")
plt.axhline(y = 2/3, color="black", label= "Probabilidade teórica com mudança de porta")
plt.ylim(0.5, 1)
plt.xlabel("Simulação", fontsize = 16)
plt.ylabel("Probabilidade de vitória", fontsize = 16)
plt.legend(fontsize = 16)
plt.show()
```
