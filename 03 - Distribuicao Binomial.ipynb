```python
from random import seed
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import binom
import math
```

```python
seed(100)
```

```python
n = 100
p = 0.3
pk = np.zeros(n)
vk = np.arange(0, n)
ns = 1000
```

```python
for j in range(0, ns):
  # success
  s = 0
  for i in range(0, n):
    r = np.random.uniform()
    if(r <= p):
      s+=1
  pk[s]+=1
# nomalization of the probability distribution
pk=pk/sum(pk)
```

```python
# making the theoerical curve
pkt = np.zeros(n+1)
vkt = np.arange(0, n+1)
for k in range(0, n+1):
  pkt[k] = (math.factorial(n)/(math.factorial(n-k)*math.factorial(k)))*(p**k)*(1-p)**(n-k)
```

$$ P(k) = \frac{n!}{(n-k)!k!} p^k (1-p)^{n-k} $$

```python
plt.figure(figsize=(10, 10))
plt.xlim(0.8 * np.min(vk[pk > 0]), 1.2 * np.max(vk[pk > 0]))
plt.bar(vk, pk, label = "Simulation")
plt.plot(vkt, pkt, 'r--', label = "Theoerical curve")

plt.xlabel("k", fontsize = 20)
plt.ylabel("P(k)", fontsize = 20)
plt.legend()
plt.show()
```
